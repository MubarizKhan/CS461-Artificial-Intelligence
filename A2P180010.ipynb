```python
'''
    --------------AI Assignment #2: Genetic Algorithm Implementation TSP----------------
    -------------- Mubariz Ahmed Khan P180010(A) ---------------------------------------   '''







import random
import numpy as np
import copy

matrix = np.array([
          [0,0,100,510,620,40,70,80,120,650], #City 1
          [60,0,60,130,40,80,90,90,440,540], #City 2
          [100,60, 0,450,450,860,910,190,10,145] , #City 3
          [510,130,450,0,70,1500, 440,220,660,250] , #City 4
          [620,40,450,70,0,260,160,330,120,50] , #City 5 onwards
          [40,80,860,1500,260,0,370,260,350,110] , #City 6 onwards
          [70,90,910,440,160,370,0,50,120,270] , #City 7 onwards
          [80,90,190,220,330,260,50,0,330,990],#City 8 onwards
          [120,440,10,660,120,350,150,330,0,330], #City 9 onwards
          [650,540,145,250,50,110,270,990,330,0]#City 10 onwards
          ])

city0 = copy.deepcopy(matrix[0])
city1 = copy.deepcopy(matrix[1])
city2 = copy.deepcopy(matrix[2])
city3 = copy.deepcopy(matrix[3])
city4 = copy.deepcopy(matrix[4])
city5 = copy.deepcopy(matrix[5])
city6 = copy.deepcopy(matrix[6])
city7 = copy.deepcopy(matrix[7])
city8 = copy.deepcopy(matrix[8])
city9 = copy.deepcopy(matrix[9])
    
city_graph = {
    'city0': city0,
    'city1': city1,
    'city2': city2,
    'city3': city3,
    'city4': city4,
    'city5': city5,
    'city6': city6,
    'city7': city7,
    'city8': city8,
    'city9': city9}

class chromosome:
  def __init__(self, dna=None):
    self.size = 10
    self.fitness = 0
    self.dna = ['city0','city1','city2','city3','city4','city5','city6','city7','city8','city9']
    # print('This is the DNA----------------', dna)
    if not dna:     
      # print('heyeyeyeyey')
      random.shuffle(self.dna)
      # print(self.dna)
      self.calculate_fitness()

    else:
      self.dna = dna
      self.calculate_fitness()
      print(self.dna,'in second argumenttt', 'This is Fitness', self.fitness)

  def calculate_fitness(self):
    
    for i in range(len(self.dna)):
      # print(len(self.dna), '<- Length ofdna .. i value -> ', i)   
      # print(self.dna, 'THIS IS DNA \n')
      # print('visiting  ', self.dna[i] )
      var1 = self.dna[i]
      # print(var1, 'this is v\r1')

      if i + 1 == len(self.dna):
        break
      var21 = self.dna[i+1]
      # print('sending', self.dna[i], '----', self.dna[i+1])
      c = int(self.get_weights(var1,var21))
      # print(c, 'value returned from get_Weight')
      self.fitness += c
      # print(self.fitness, 'this is fitness \n')
    return self.fitness

  def str_to_digit(self,str1):
    digit = int(str1[4:])
    return digit

  def get_weights(self,list1,list2):
    str_splitt = self.str_to_digit(list2) #Index of array to be accessed
    dict_array1 = city_graph[list1]
    weight = dict_array1[str_splitt]
    
    # print('this is', list1,'->', list2, 'Weight is:', weight )
    return weight




    

  
```

```python
import copy

class GeneticAlgo:
  def __init__(self, pool_size = 1000):
    self.pool_size = pool_size
    self.pool = []
    self.chromosome = chromosome()
    # print(len(self.pool))
    self.selection_list = []
    self.pool_fitness_log = []
    self.goal = chromosome()

    self.initialize_population()
    self.avg_pool_fitness = self.pool_fitness()
    # print(self.avg_pool_fitness, 'average fitness of pool')  
    self.start_simulation()
    self.evaluation()
    self.selection_criteria()
    # print(self.initial_selection())
    # print(self.pool_fitness(), 'this is the average fitness of pool')
    
    print(self.goal.fitness)
    self.print_goal()
    # self.generate_population()
  
  def print_goal(self):
    print('-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/')
    print('-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/')
    print('-/-/-/---GOAL REACHED IN THIS SIMULATION---/-/-/-/-/-/-/-/-/-/-/')
    print('\n\n GOAL: ', self.goal.dna, 'Fitness: ', self.goal.fitness)
    print('-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/')
    print('-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/-/')
    
    return self.pool
  def initialize_population(self):
    for i in range(self.pool_size):
      chromo = chromosome()
      print('chromo s DNA ', chromo.dna)
      self.pool.append(chromo)
      print(self.pool[i].fitness , 'Fitness of current DNA '   )

    '''
    selection doctstring---------------
      fitness_value -- the minimum the better -----
      population -> 100 -------------- 4, p1,p2 -> c1,c2:  p1,p3-> c3,c4. p1,p4-> C5,C6
                                        P2,P3,-> C7,C8: P2,P4-> C9,C10
                                        P3,P4 -> C11, C12
      parents -> f values

      min(fitness/total fitness)

      Implementation: 

      Step 2:Selection of Parents
      1.Parents will be selected random from the population
       in the first step.
       2.High fitness value population will have high chances of selection as 
       a parenti.e. 
       - best 4 nodes will be selected for breeding
       if current_node.fitness < avg_poo_fitness:
         append current_node in pool
        when time period of generation ends, remove that certain generation from node
    '''
  def poolFitness_track(self, var):
    pool_fitness_log = []
    pool_fitness_log.append(var)
    return pool_fitness_log 

  def pool_fitness(self):
    '''calculate fitness of pool, and then calculate fitness/selection probability of 
      each node for selection of parents, the higher the probability the lesser the chances
      of mating, as we are interested in the shortest combination of cities-TSP
    '''
    total_pool_fitness = 0
    avg_pool_fitness = 0

    for i in range(len(self.pool)):
      var = self.pool[i].fitness
      total_pool_fitness += var 
    avg_pool_fitness = (total_pool_fitness / len(self.pool) )

    self.pool_fitness_log = self.poolFitness_track(avg_pool_fitness)

    return avg_pool_fitness

  def cross_send(self,chromosome1,chromosome2):
    
    self.selection_list.append(chromosome1)
    self.selection_list.append(chromosome2)
    self.evaluation()

    return self.selection_list

  def evaluation(self):
    # print(len(self.pool), 'pool size -----\\->', self.avg_pool_fitness )
    # print ( len(self.selection_list) , 'length of selection list')

    for i in self.selection_list:
      # print('1.selection list values, i = ', i.fitness, 'avg ', self.avg_pool_fitness)

      if int(i.fitness) < int(self.avg_pool_fitness):
        print('2.in loop fitness->', int(i.fitness), '----- avg_pool_fitness ->', int(self.avg_pool_fitness) )
        print('selction list value is lesser--- i.fit->', i.fitness, '--', self.avg_pool_fitness , '<- avg pool fitness')
    
        # print(i.fitness, 'These have been selected')
        self.selection_list.remove(i)
        self.pool.append(i)
      else:
    
          self.selection_list.remove(i)

      self.avg_pool_fitness = self.pool_fitness()

    print(len(self.pool), 'pool size')
    

  def breeding(self, chromosome1, chromosome2):
    self.avg_pool_fitness = self.pool_fitness()
    var =  self.crossover(chromosome1, chromosome2 )

    print(len(self.pool), 'This is pool after ', self.avg_pool_fitness )
    


  def mutation(self, chromosome):
    index1 = random.randint(2, 8)
    index2 = random.randint(3,7)

    print(index1,'<-index1 -To be Mutated- index2->', index2)
    print(chromosome.dna)
    print('chromosome[index1]-->', chromosome.dna[index1] )
    print('chromosome[index2]-->', chromosome.dna[index2] )

    mut_index1 = chromosome.dna[index1]
    temp = chromosome.dna[index1]
    mut_index2 = chromosome.dna[index2]
    chromosome.dna[index1] = mut_index2
    chromosome.dna[index2] = temp   

    return chromosome


  def selection_criteria(self):
    min_fitness1 = self.pool[0].fitness
    min_value_node1 = self.pool[0]

    min_fitness2 = self.pool[0].fitness
    min_value_node2 = self.pool[0]

    min_fitness3 = self.pool[0].fitness
    min_value_node3 = self.pool[0]

    min_fitness4 = self.pool[0].fitness
    min_value_node4 = self.pool[0]

    for i in self.pool:

      if (min_fitness1 >= i.fitness ):
        min_fitness4 = min_fitness3
        min_fitness3 = min_fitness2
        min_fitness2 = min_fitness1
        min_fitness1 = i.fitness

        min_value_node4 = copy.deepcopy(min_value_node3)
        min_value_node3 = copy.deepcopy(min_value_node2)
        min_value_node2 = copy.deepcopy(min_value_node1)
        min_value_node1 = copy.deepcopy(i)

      elif min_fitness2 >= i.fitness:
        min_fitness4 = min_fitness3
        min_fitness3 = min_fitness2
        min_fitness2 - i.fitness

        min_value_node4 = copy.deepcopy(min_value_node3)
        min_value_node3 = copy.deepcopy(min_value_node2)
        min_value_node2 = copy.deepcopy(i)
        
      elif min_fitness3 >= i.fitness:
        min_fitness4 = min_fitness3
        min_fitness3 = i.fitness

        min_value_node4 = copy.deepcopy(min_value_node3)
        min_value_node3 = copy.deepcopy(i)

      elif min_fitness4 >= i.fitness:
        min_fitness4 = i.fitness 
        min_value_node4 = copy.deepcopy(i)
      
    self.goal = min_value_node1

    self.breeding(min_value_node1, min_value_node2) 
    self.breeding (min_value_node3,min_value_node4)
    print(self.goal.fitness,'This is minimum ---------- goal to be reached')


  def start_simulation(self):
    self.avg_pool_fitness = self.pool_fitness()
    #0,1 1,2 2,3 -> 2
    var = []
    generation = []
    for i in range(self.pool_size):
      generation.append(self.pool[i])
      if i + 1 == len(self.pool):
        break
      if i % 10 == 0:
        discard = generation.pop(0)
        print('Theme to discard old variables', discard)
        self.pool.remove(discard)   
      var =  self.crossover(self.pool[i], self.pool[i+1] )
    return var
  
  def crossover(self, chromosome1, chromosome2):
    print("---------------------------Crossover-----------------------------")
    print("---------------------------Mutation-----------------------------")
    divid = random.randint(2, 8)
    # print('index', divid)
    print("--selected Parents-- \n")

    print("parent1:  ", chromosome1.dna,"\n")
    print("parent2:  ", chromosome2.dna,"\n")

    left_of_chromosome1 = chromosome1.dna[:divid] # left half -> c1
    left_of_chromosome2 = chromosome2.dna[:divid]
    right_of_chromosome1 = chromosome1.dna[divid:]#right half -> c1
    right_of_chromosome2 = chromosome2.dna[divid:]

    child1 = left_of_chromosome1 + right_of_chromosome2
    child2 = left_of_chromosome2 + right_of_chromosome1


    print(left_of_chromosome1[:divid], 'left_of_chromosome1', left_of_chromosome2,'left_of_chromosome2')
    print(right_of_chromosome1,'right_of_chromosome1', right_of_chromosome2,'right_of_chromosome2')
    print(right_of_chromosome2,'right_of_chromosome2')
   
    cr1 = chromosome(child1)
    cr2 = chromosome(child2)

    print("After Crossover New Childs \n child1: ", cr1.dna, "\n child2: ",cr2.dna)
    print('This is fitness of children: ', cr1.fitness, cr2.fitness)
    
    cr1 = self.mutation(cr1)
    cr2 = self.mutation(cr2)

    return self.cross_send(cr1,cr2)


```
